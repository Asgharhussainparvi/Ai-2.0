```python
# Extract time-based features
stock_df['Year'] = stock_df['Date'].dt.year
stock_df['Month'] = stock_df['Date'].dt.month
stock_df['Day'] = stock_df['Date'].dt.day
stock_df['Weekday'] = stock_df['Date'].dt.weekday 
```

```python
# PDF of all
plt.figure(figsize=(6, 4))
sns.kdeplot(stock_df['Price'], fill=True, color='blue', label='Price')
sns.kdeplot(stock_df['High'], fill=True, color='red', label='High')
sns.kdeplot(stock_df['Low'], fill=True, color='green', label='Low')

plt.title('Probability Density Function of USD price', fontsize=14)
plt.xlabel('Score')
plt.ylabel('Density')
plt.xticks(np.arange(20, 300, 50))
plt.legend()
plt.show()
```

```python
# Calculate mean and standard deviation
mean_math = stock_df['Price'].mean()
std_math = stock_df['Price'].std()
print(f"Mean: {mean_math}, Standard Deviation: {std_math}")

# Generate x values for Gaussian curve
x = np.linspace(stock_df['Price'].min(), stock_df['Price'].max(), 100)

# Compute Gaussian distribution
gaussian = (1 / (std_math * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_math) / std_math) ** 2)
plt.figure(figsize=(7, 4))
plt.plot(x, gaussian, color='red', linewidth=2, label='Gaussian Curve')
plt.title('Gaussian Distribution of Price', fontsize=14)
plt.xlabel('Math Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```
