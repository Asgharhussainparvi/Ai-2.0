**Data Cleaning**

```python
car_data.replace(["?","$","%","@"],np.nan,inplace=True)
car_data.head()
```

```python
car_data.replace(["?"," ","/",",","_","-"],np.nan,inplace=True)
car_data.head()
```

```python
car_data.drop(columns="Unnamed: 0",inplace=True)
car_data.head()
```

```python
car_data.isnull().sum()
```

```python
car_data.dtypes
```

```python
avg_normal_loss = car_data["normalized-losses"].astype("float").mean()
print(f"Average of normal losses is :{avg_normal_loss}")
```

```python
car_data["normalized-losses"].replace(np.nan,avg_normal_loss,inplace=True)
car_data.head()
```

```python
avg_stroke = car_data["stroke"].astype("float").mean()
print(avg_stroke)
car_data["stroke"].replace(np.nan,avg_stroke,inplace=True)
car_data.stroke
```

```python
avg_bore = car_data["bore"].astype("float").mean()
print(avg_bore)
car_data["bore"].replace(np.nan,avg_bore,inplace=True)
car_data.bore
```

```python
avg_horsepower = car_data["horsepower"].astype("float").mean()
print(avg_horsepower)
car_data["horsepower"].replace(np.nan,avg_horsepower,inplace=True)
car_data.horsepower
```

```python
avg_peakrpm = car_data["peak-rpm"].astype("float").mean()

print(avg_peakrpm)
car_data["bore"].replace(np.nan,avg_peakrpm,inplace=True)

car_data["peak-rpm"]
```

```python
car_data.dropna(subset=['price'],axis = 0,inplace=True)
car_data.reset_index(drop=True,inplace=True)
car_data.head(20)
```

```python
car_data[["bore","stroke","price","peak-rpm"]] = car_data[["bore","stroke","price","peak-rpm"]].astype("float")
car_data["normalized-losses"] = car_data["normalized-losses"].astype("int")
```
