```python
average_pollutants = df.groupby('Country Label')['Value'].mean().reset_index()
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(data=average_pollutants, x='Country Label', y='Value')
plt.title('Average Pollutant Levels by Country')
plt.xlabel('Country')
plt.ylabel('Average Pollutant Level (µg/m³)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
top_10_countries = average_pollutants.sort_values('Value', ascending=False).head(10)
```

```python
name = top_10_countries['Country Label']

color=['orange','grey','purple','yellow','blue','green','red','pink','brown','cyan']
explode=(0,0,0,0,0.02,0,0,0,0,0)
size=[df['Value'][df['Country Label']==i].mean() for i in name]
plt.pie(size, labels=name, autopct='%1.2f%%', colors=color, explode=explode)
plt.show()
```

```python
# Create categories
def categorize_value(value):
    if value <= 50:
        return 'Hazardous'
    elif 51 <= value <= 100:
        return 'Unhealthy'
    elif 101 <= value <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif 151 <= value <= 200:
        return 'Unhealthy'
    elif 201 <= value <= 300:
        return 'Moderate'
    else:
        return 'Good'
```

```python
df['value_Category'] = df['Value'].apply(categorize_value)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='value_Category', palette='viridis')
plt.title('Distribution of Value Categories')
plt.xlabel('Value Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Last Updated'])
df.set_index('Date', inplace=True)
yearly_avg = df['Value'].resample('M').mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(yearly_avg.index, yearly_avg.values, linestyle='-', color='b')
plt.title('Yearly Average Pollutant Levels')
plt.xlabel('Date')
plt.ylabel('Average Pollutant Level (µg/m³)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
