### **Creating Time-Based Features**
- Extract **year**, **month**, **day**, and **weekday** from the `Order Date`.
- These features can help analyze trends over time.



```python
df["Order Date"]
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date']
```

```python
# Extract time-based features
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day
df['Order Weekday'] = df['Order Date'].dt.weekday  # Monday=0, Sunday=6
```

```python
monthly_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
monthly_sales
```

```python
# Line plot for monthly sales trend

sns.lineplot(x='Order Month', y='Sales', hue='Order Year', data=monthly_sales, marker='o')
plt.title('Monthly Sales Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

### **Calculating Profit Margin**
- Profit Margin = (Profit / Sales) * 100
- This feature shows the profitability of each order.



```python
# Calculate Profit Margin
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
```

```python
# Profit Margin Distribution
sns.histplot(df['Profit Margin'], bins=30, color='purple')
plt.title('Distribution of Profit Margin')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Frequency')
plt.show()
```

### **Categorizing Discounts**
- Create discount levels: **Low**, **Medium**, and **High**.
- This helps analyze the impact of discounts on sales and profit.

```python
df.Discount.describe()
```

```python
# Categorize Discounts
df['Discount Level'] = pd.cut(df['Discount'], bins=[-1, 0.1, 0.3, 1], labels=['Low', 'Medium', 'High'])
df['Discount Level']
```

```python
# Discount Level vs. Profit
sns.boxplot(x='Discount Level', y='Profit', data=df, palette='coolwarm')
plt.title('Profit by Discount Level')
plt.xlabel('Discount Level')
plt.ylabel('Profit')
plt.show()
```

### **Calculating Shipping Duration**
- Shipping Duration = Ship Date - Order Date
- This feature helps analyze delivery efficiency.

```python
# Convert 'Ship Date' to datetime format
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Calculate Shipping Duration
df['Shipping Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df["Shipping Duration"].describe()
```

```python
# Shipping Duration Distribution
sns.histplot(df['Shipping Duration'], bins=30, kde=True, color='orange')
plt.title('Distribution of Shipping Duration')
plt.xlabel('Shipping Duration (Days)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Segment'].value_counts()
```

```python
# Total and Average Profit by Segment
df['Total Profit by Segment'] = df.groupby('Segment')['Profit'].transform('sum')
df['Total Profit by Segment']
```

```python
df['Avg Profit by Segment'] = df.groupby('Segment')['Profit'].transform('mean')
df['Avg Profit by Segment']
```

```python
# Sales Contribution by Category
total_sales = df['Sales'].sum()
df['Sales Contribution'] = (df.groupby('Category')['Sales'].transform('sum') / total_sales) * 100
```

```python
# Display the new features
df['Sales Contribution']
```

```python
# Sales Contribution by Category
sns.barplot(x='Category', y='Sales Contribution', data=df, palette='viridis')
plt.title('Sales Contribution by Category')
plt.xlabel('Category')
plt.ylabel('Sales Contribution (%)')
plt.show()
```
